## Generating Sound

The sound of the Aberrants is so unlike that of humans that we may need to learn more about different types of sound waves, in order to gain insight into what these creatures sound like. Thankfully, the Eldritch tongue allows you to magically generate different types of sound, combine or modify them however you like, and listen to the results.

Remember that the Eldritch tongue represents sound as a simple list of numbers. You can write down any list you like, and listen to the results.

```python
my_list = [-1, -5, 1, 2, 3, -1, 4, 3]
listen(my_list)
```

Ah, but this list was much too short. Most sound lists have thousands of numbers in them, so we will need to make our example bigger in order to hear it.

In the Eldritch tongue, you can __repeat__ a list using the multiplication operator *.

```python
print 3*my_list
```

This has made one big list by copying the elements in `my_list` three times.

__Challenge__. Make a super long list by repeating `my_list` ten thousand times, and listen to the results.
